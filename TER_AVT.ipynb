```python
# look only at a subset of edit values
melt1 = melt[(melt.edit == "---") | 
             (melt.edit == "S") | 
             (melt.edit == "H") | 
             (melt.edit == "D") | 
             (melt.edit == "SS")]
#melt1 = melt
melt1 =  melt1[(melt1.Task == "T") | (melt1.Task == "P") ]
melt1 =  melt1[(melt1.Dur < 5000) & (melt1.Dur > 0) ]

# Ordinary Least Squares (OLS) model 
model = ols('Dur ~ edit', data=melt1).fit()
print(model.summary())
```

## Barplot for different hypothesis (same MT, other MT, HT)

```python
# re-size df into 2 columns with header --> variable - editOperation --> value 
melt = pd.melt(refST.reset_index(), value_vars=["TERedO", "TERedP01","TERedP02","TERedRP06","TERedRP12"])
melt1 = melt[(melt.value == "S") | (melt.value == "H") | (melt.value == "D") | (melt.value == "SS")]

print(f"\nmelted table {melt.shape} -- {melt1.shape}: \n {melt1.head()}")

ct = pd.crosstab(melt1.value, melt1.variable, margins = False)
print(f"\ncontingency table {ct.shape}:\n {ct.head(10)}\n")

stacked = ct.stack().reset_index()

stacked.columns = ['edit', 'hyp', 'count']
stacked["count"] = pd.to_numeric(stacked["count"])
stacked["logCount"] = np.log(stacked["count"])

# sort df accourding to counts
stacked = stacked.sort_values(by ='count', ascending = False)
stacked1 = stacked[(stacked["count"] > 10)]
print(f"\nstacked table {stacked.shape} deduced to {stacked1.shape}:\n {stacked1.head(10)}\n")

sns.barplot(x=stacked1["edit"], y=stacked1["logCount"], hue=stacked1["hyp"])
```

```python
import matplotlib.pyplot as plt
#print(melt.head())


# re-size df into 2 columns with header --> variable - editOperation --> value 
melt = pd.melt(refST.reset_index(), id_vars=["Task","Dur","HTra"], value_vars=["TERedO"])

print(melt.head())

# ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']
#marker = {"I":"o","D":".","S":",","M":"x"}
marker = {"I":"red","D":"green","S":"blue","H":"black"}
marker = {"H":"black"}
print (marker.keys())
D = []
H = []
M = []
L = []

for index, row in melt.iterrows(): 
    if(row.value in marker.keys() and row.Dur > 0 and row.Dur < 3000) :
        D.append(row.Dur)
        H.append(row.HTra)
        M.append(marker[row.value])

plt.scatter(D, H, s=2, c=M, alpha=0.3)

# Set the line chart title and the text font size.
plt.title("Relation Dur - Htra for different edit operations", fontsize=14)

# Set x axes label.
plt.xlabel("Duration", fontsize=12)

# Set y axes label.
plt.ylabel("Tra", fontsize=12)
```
